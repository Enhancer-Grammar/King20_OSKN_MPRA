```python
from bin.King20 import king20, tfbs_encode, ResidualBind
```

```python
data = king20("SYN", "./processed/2024_03_10")
data = data[data["Element_id"] != "Basal"]
data
```

```python
# Check for NaNs in "Sequence" and remove them
data["Sequence"].isna().sum()
data = data.dropna(subset=["Sequence"])
data
```

```python
# Add function registry for normalization
data["log_CRE_normalized_mean"] = np.log2(data["CRE_normalized_mean"])
```

```python
# 
nan_mask = np.isnan(data["log_CRE_normalized_mean"])
data = data[~nan_mask]
data
```

```python
seqs = data["Sequence"].values
patterns = dict(
    sox2 = ["GCTCATTGTTTC"],
    pou5f1 = ["GGGATGCTAATC"],
    esrrb = ["GTTCAAGGTCAC"],
    klf4 = ["GGGGTGGGGCCG"],
)
include_rc = True
```

```python
# Encode
ohe_in_seqs_tensor, tfs = tfbs_encode(
    seqs=seqs,
    patterns=patterns,
    include_rc=include_rc,
    max_length=80,
    verbose=False,
)
```

```python
sdata = xr.Dataset({
    "ohe_seq": xr.DataArray(ohe_in_seqs_tensor, dims=["_sequence", "_ohe", "_length"]),
    #"ohe_seq": xr.DataArray(sp.ohe(sp.pad_seqs(data["Sequence"], length=80, pad="both", pad_value="N"), sp.alphabets.DNA).transpose(0, 2, 1), dims=["_sequence", "_ohe", "_length"]),
    "target": xr.DataArray(data["log_CRE_normalized_mean"].values, dims=["_sequence"]),
})
sdata
```

```python
test_seqs = torch.tensor(sdata["ohe_seq"].isel(_sequence=slice(0, 10)).values, dtype=torch.float32)
test_seqs
```

# Training module

```python
from seqmodels import Module
```

```python
import torch
```

```python
test_dict = {"ohe_seq": torch.tensor(sdata["ohe_seq"].isel(_sequence=slice(0, 10)).values, dtype=torch.float32)}
```

```python
# Make a module
module = Module(
    arch=arch,
    input_vars=["ohe_seq"],
    output_vars=["output"],
    target_vars=["target"],
    loss_fxn="mse",
    train_metrics_fxn=["r2", "pearson", "spearman"],
    val_metrics_fxn=["r2", "pearson", "spearman"],
    scheduler="reduce_lr_on_plateau",
)
module, module(test_dict).shape
```

# DataLoaders

```python
# Split into train and validation based on random 50/50
train_idx, val_idx = np.split(np.random.permutation(len(sdata["_sequence"])), [int(0.5 * len(sdata["_sequence"]))])
train_sdata = sdata.isel(_sequence=train_idx)
valid_sdata = sdata.isel(_sequence=val_idx)
train_sdata, valid_sdata
```

```python
train_data = ohe_in_seqs_tensor[train_idx]
valid_sdata = sdata.isel(_sequence=val_idx)
train_sdata, valid_sdata
```

```python
import seqdata as sd
```

```python
# Train dataloader
train_dl = sd.get_torch_dataloader(
    train_sdata.load(),
    sample_dims="_sequence",
    variables=["ohe_seq", "target"],
    batch_size=32,
    shuffle=True,
    num_workers=0,
    drop_last=True,
    pin_memory=True,
)
batch = next(iter(train_dl))
batch["ohe_seq"].shape, batch["target"].shape
```

```python
# Validation dataloader
valid_dl = sd.get_torch_dataloader(
    valid_sdata.load(),
    sample_dims="_sequence",
    variables=["ohe_seq", "target"],
    batch_size=32,
    shuffle=False,
    num_workers=0,
    drop_last=False,
    pin_memory=True,
)
batch = next(iter(valid_dl))
batch["ohe_seq"].shape, batch["target"].shape
```

# Trainer

```python
from pytorch_lightning import Trainer
```

```python
# Logger
from pytorch_lightning.loggers import CSVLogger
logger = CSVLogger(save_dir="log", name="", version="")
```

```python
# Add ModelCheckpoint, EarlyStopping and LearningRateMonitor callbacks
from pytorch_lightning.callbacks import ModelCheckpoint, EarlyStopping, LearningRateMonitor
callbacks = [
ModelCheckpoint(
    dirpath=os.path.join(
        logger.save_dir, 
        logger.name, 
        logger.version, 
        "checkpoints"
    ),
    save_top_k=5,
    monitor="val_loss_epoch",
    mode="min",
),
    EarlyStopping(
        monitor="val_loss_epoch",
        patience=10,
        mode="min",
    ),
    LearningRateMonitor(),
]
```

```python
# Trainer
trainer = Trainer(
    logger=logger,
    callbacks=callbacks,
    max_epochs=100,
    check_val_every_n_epoch=1,
)
```

# Fit

```python
# Fit the weigths
trainer.fit(module, train_dl, valid_dl)
```

```python
# Get the best model weights
best_model_path = trainer.checkpoint_callback.best_model_path
copy_path = os.path.join("best_model.ckpt")
os.system(f"cp {best_model_path} {copy_path}")
```
